```python
tinggi = float(input("Masukkan tinggi badan (meter): "))
bmi = float(input("Masukkan BMI yang diharapkan: "))

berat = bmi * (tinggi ** 2)

print("Berat badan yang diperlukan adalah %.2f kg" % berat)
```

```python
x = int(input("Masukkan nilai x (bilangan bulat): "))

if x != 0:
    hasil = 2*(x**3) + 2*x + 15/x
    print("Nilai f(x) adalah:", hasil)
else:
    print("x tidak boleh 0 karena terjadi pembagian dengan nol.")
```

```python
upah = float(input("Masukkan gaji per jam: "))
jam_kerja = float(input("Masukkan jumlah jam kerja per minggu: "))

lama_kerja = 5

# Total pendapatan sebelum pajak
total_kotor = upah * jam_kerja * lama_kerja

# Pajak 14%
potongan_pajak = 0.14 * total_kotor
total_bersih = total_kotor - potongan_pajak

# Pengeluaran setelah pajak
belanja_pakaian = 0.10 * total_bersih
belanja_atk = 0.01 * total_bersih

# Sisa uang
uang_tersisa = total_bersih - belanja_pakaian - belanja_atk

# Sedekah 25% dari sisa uang
total_sedekah = 0.25 * uang_tersisa

# Pembagian sedekah
bagian_yatim = 0.30 * total_sedekah
bagian_dhuafa = 0.70 * total_sedekah

# Output
print("1. Pendapatan sebelum pajak: Rp %.2f" % total_kotor)
print("2. Pendapatan setelah pajak: Rp %.2f" % total_bersih)
print("3. Uang untuk pakaian dan aksesoris: Rp %.2f" % belanja_pakaian)
print("4. Uang untuk alat tulis: Rp %.2f" % belanja_atk)
print("5. Jumlah uang yang disedekahkan: Rp %.2f" % total_sedekah)
print("6. Uang untuk anak yatim: Rp %.2f" % bagian_yatim)
print("7. Uang untuk kaum dhuafa: Rp %.2f" % bagian_dhuafa)
```
