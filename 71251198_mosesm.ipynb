```python
def bandingkan_file(file1, file2):
    try:
        with open(file1, "r", encoding="utf-8") as f1:
            teks1 = f1.readlines()

        with open(file2, "r", encoding="utf-8") as f2:
            teks2 = f2.readlines()

        jumlah_baris = max(len(teks1), len(teks2))
        ada_perbedaan = False

        for i in range(jumlah_baris):
            baris1 = teks1[i].rstrip("\n") if i < len(teks1) else "[Baris tidak ada]"
            baris2 = teks2[i].rstrip("\n") if i < len(teks2) else "[Baris tidak ada]"

            if baris1 != baris2:
                ada_perbedaan = True
                print(f"Perbedaan pada baris ke-{i + 1}:")
                print(f"File 1: {baris1}")
                print(f"File 2: {baris2}")
                print("-" * 40)

        if not ada_perbedaan:
            print("Kedua file sama, tidak ada perbedaan.")

    except FileNotFoundError:
        print("Salah satu file tidak ditemukan.")
    except Exception as e:
        print(f"Terjadi kesalahan: {e}")


file_pertama = input("Masukkan nama file pertama: ")
file_kedua = input("Masukkan nama file kedua: ")

bandingkan_file(file_pertama, file_kedua)
```

```python
isi_soal = """1+1 = || 2
Bendera Indonesia? || Merah Putih
Kota gudeg adalah: || Yogyakarta
Komponen PC untuk penyimpanan file adalah... || harddisk
50 * 20 = || 1000
"""

with open("soal.txt", "w", encoding="utf-8") as file:
    file.write(isi_soal)

print("File soal.txt berhasil dibuat")
```

```python
nama_file = "soal.txt"
print("nama file1:", nama_file)

try:
    with open(nama_file, "r", encoding="utf-8") as file:
        for baris in file:
            if "||" in baris:
                soal, kunci = baris.strip().split("||")

                soal = soal.strip()
                kunci = kunci.strip().lower()

                print(soal)
                jawaban = input("Jawab: ").strip().lower()

                if jawaban == kunci:
                    print("Jawaban benar!\n")
                else:
                    print("Jawaban salah!\n")

except FileNotFoundError:
    print("File soal.txt tidak ditemukan.")
```
