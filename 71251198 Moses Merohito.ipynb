```python
def cek_angka(a, b, c):
    a = int(input("Masukan angka 1"))
    b = int(input("Masukan angka 2"))
    c = int(input("Masukan angka 3"))

    if a != b and a != c and b != c:
        if a + b == c or a + c == b or b + c == a:
            return True
    return False
print(cek_angka(0,0,0))

```

```python
def cek_digit_belakang(a, b, c):
    d1 = a % 10
    d2 = b % 10
    d3 = c % 10
    
    if d1 == d2 or d1 == d3 or d2 == d3:
        return True
    else:
        return False

a = int(input("Masukkan angka pertama: "))
b = int(input("Masukkan angka kedua: "))
c = int(input("Masukkan angka ketiga: "))

print(cek_digit_belakang(a, b, c))
```

```python
celcius_ke_fahrenheit = lambda C: (9/5) * C + 32
celcius_ke_reamur = lambda C: 0.8 * C

print("Input C = 100 → F =", celcius_ke_fahrenheit(100))
print("Input C = 80 → R =", celcius_ke_reamur(80))
print("Input C = 0 → F =", celcius_ke_fahrenheit(0))
```
